```python
#Find customers with credit history greater than 10 years but credit score less than 600.
df[(df.credit_history_years > 10)&(df.credit_score < 600)]
```

```python
#Among Self-Employed customers, show those whose loan was rejected.
df[(df.occupation_status == 'Self-Employed')&(df.loan_status == 0)]
```

```python
#Display customers with: • no defaults • no derogatory marks • credit score above 650
df[(df.defaults_on_file == 0)&(df.derogatory_marks == 0)&(df.credit_score > 650)]
```

```python
#Identify “high-risk customers” where: • credit score < 600 • OR debt-to-income ratio > 0.6 • OR delinquencies in last 2 years ≥ 2
df[(df.credit_score < 600)|(df.debt_to_income_ratio > 0.6)|(df.delinquencies_last_2yrs > 2)]
```

```python
#Show approved loans where loan-to-income ratio is greater than 1.
df[(df.loan_status == 1)&(df.loan_to_income_ratio > 1)]
```

```python
#Display customers whose savings_assets are less than 500 and current_debt is greater than 20,000.
df[(df.savings_assets < 500)&(df.current_debt > 20000)]
```

```python
#Find customers who are either Students OR have years_employed less than 2 years.
df[(df.occupation_status == 'Student')&(df.years_employed < 2)]
```

```python
#Show customers with interest rate between 10% and 15%.
df[(df.interest_rate >= 10)&(df.interest_rate <= 15)]
```

```python
#Display customers who applied for Debt Consolidation and have loan amount greater than 20,000.
df[(df.loan_intent == 'Debt Consolidation')&(df.loan_amount > 20000)]
```

```python
#Find customers whose annual income is less than 30,000 but loan amount is greater than 
df[(df.annual_income < 30000)&(df.loan_amount > 30000)]
```
